### Le cercle
Référence - Panoramath - Secondaire 2

cercle

```python
import math
class Cercle:
    def __init__(self, rayon, diametre, perimetre, aire, angle_secteur, aire_secteur):
        
        if rayon != None:
            self.rayon = rayon
            self.diametre = rayon * 2

        if diametre != None:
            self.diametre = diametre
            self.rayon = round(diametre / 2)

        if aire != None:
            self.aire = aire
            self.rayon = round(math.sqrt(aire / math.pi), 2)

        if perimetre != None:
            self.perimetre = perimetre
            self.rayon = round(perimetre / (2 * math.pi), 2)

        self.diametre = self.rayon * 2
        
        if perimetre == None:
            self.perimetre = self.perimetre_cercle()
        
        if aire == None:
            self.aire = self.aire_cercle()

        if angle_secteur != None:
            self.angle_secteur = angle_secteur
            self.aire_secteur = round(self.aire_cercle() * angle_secteur / 360, 2)
        
        if aire_secteur != None:
            self.aire_secteur = aire_secteur
            self.angle_secteur = round(aire_secteur * 360 / self.aire_cercle(), 2)
        
    def aire_cercle(self):
        return round(math.pi * self.rayon ** 2, 2)

    def perimetre_cercle(self):
        return round(2 * math.pi * self.rayon, 2)

    def angle_secteur_cercle(self):
        return self.angle_secteur

    def aire_secteur_cercle(self):
        return self.aire_secteur

    def afficher_info(self):
        return f"{cercle.rayon:<10}{cercle.diametre:<10}{cercle.perimetre:<10}{cercle.aire:<10}{cercle.angle_secteur:<10}{cercle.aire_secteur:<10}"


cercle = Cercle(5, None, None, None, 86, None)
print(cercle.afficher_info())

cercle = Cercle(None, 126, None, None, 50, None)
print(cercle.afficher_info())

cercle = Cercle(None, None, None, 365, None, 200)
print(cercle.afficher_info())    
    
cercle = Cercle(None, None, 1350, None, None, 12653)
print(cercle.afficher_info())

cercle = Cercle(12, None, None, None, 80, None)
print(cercle.afficher_info())

cercle = Cercle(None, None, None, 4000, None, 800)
print(cercle.afficher_info())

```
