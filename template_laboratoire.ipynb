#### Question 1  
Créer une fonction ``liste_cubes(nombre)`` qui prend en entrée un paramètre ``nombre``, et qui retourne la liste des cubes des entiers compris entre $1$ et $nombre$. Par exemple, pour $nombre = 5$, on obtient la liste ``[1, 8, 27, 64, 125]``.

```python
# Réponse à la question 1
def liste_cubes(nombre):
    liste = []
    for i in range(1, nombre+1):
        liste.append(i**3)
    return liste

n = 5
liste_cubes(n)
print(f"La liste des cubes de 1 à {n} est {liste_cubes(n)}")

# Une autre solution à la question 1
def liste_cubes_v2(nombre):
    return [i**3 for i in range(1, nombre+1)]

n = 7
liste_cubes_v2(n)
print(f"La liste des cubes de 1 à {n} est {liste_cubes_v2(n)}")
```
