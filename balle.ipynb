# La hauteur d'une balle

Calculer et afficher la hauteur d'une balle lancée du sommet d'un immeuble après un temps $t$ de 4 secondes si sa position verticale en mètre est donnée par :
$$h(t) = -4,9t^2 + 10t + 50$$

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 5, 100)

plt.plot(t, -4.9*t**2 + 10*t + 50)
t = 4
h = -4.9*t**2 + 10*t + 50
plt.plot(t, h, color='red', marker='o', markersize=10)
plt.axhline(y = 0, color = 'r', linestyle = '-') 

plt.xlabel("t (secondes)")
plt.ylabel("hauteur (mètres)")
plt.show()

print(f"Au temps t = 4 secondes, la hauteur est de {h} mètres.")
```
