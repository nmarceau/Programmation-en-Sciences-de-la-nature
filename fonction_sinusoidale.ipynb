$$f(x) = a\times sin(b(x−h)) + k$$


- Le paramètre $a$ est lié à l'amplitude.
- Le paramètre $b$ est lié à la période.
- Le paramètre $h$ est lié au déphasage.
- Le paramètre $k$ est lié à l'axe d'oscillation.

### La grande roue
#### Une résolution de problèmes impliquant la fonction sinus

Dans certaines villes une grande roue est installée afin d'offrir une vue panoramique aux touristes qui le désirent.

Pour grimper à l'intérieur d'un tel manège, une plateforme a été installée à une hauteur équivalente à la moitié de celle de la grande roue. Afin d'éviter toutes collisions avec les passants, le point le plus bas de la grande roue se situe à 5 m du sol. Une fois le sommet atteint, les usagers se retrouvent à une hauteur impressionnante de 131 m. Dès l'embarquement fait, la nacelle dans laquelle les usagers prennent place se dirige vers le bas.

Sachant que la vue devient particulièrement spectaculaire à partir de 120 m d'altitude et que la grande roue prend 32 minutes pour effectuer un tour complet, pendant combien de temps les touristes seront-ils impressionnés par le paysage?

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import sympy as sp

min = 5
max = 131
 
a = (max-min) / 2 * -1
b = math.pi/16
h = 0
k = (max-min) / 2 + min

x = np.linspace(0, 50, 1000)
y = a * np.sin(b * (x-h)) + k 

print("Valeur maximale de la fonction sinusoidale", np.max(y))
print("Valeur minimale de la fonction sinusoidale", np.min(y))

legend = f'{a} * sin({b} * (x - {h}) + {k})'
plt.plot(x, y, label=legend, color='red', linestyle='dashed', linewidth=2)
plt.title('Graphique de la fonction sinusoïdale')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.plot(0, k, 'ro', label=f'Départ ({h}, {k})')
plt.axhline( y = 5, color = 'blue', linestyle = 'dashed', linewidth = 2, label = 'y = 5')
plt.axhline( y = 131, color = 'blue', linestyle = 'dashed', linewidth = 2, label = 'y = 131')

plt.legend()
plt.grid(True) 
plt.show() 

x = sp.symbols('x')
f = a * sp.sin(b * (x-h)) + k 
equation = sp.Eq(f, 120 )
solution = sp.solve(equation, x)
print(solution)
```
