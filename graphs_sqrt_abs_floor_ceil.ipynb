### La racine carr√©e

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100) 
y = np.sqrt(x) 

plt.plot(x, y, label='y = ‚àöx')  
plt.xlabel('x')  
plt.ylabel('‚àöx') 
plt.title('Graphique de la racine carr√©e de x') 
plt.legend()  
plt.grid(True)  
plt.show()  
```

### La valeur absolue

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

plt.plot(x, abs(x) - 2)
plt.axhline(y = 0, color = 'black', linestyle='--') 
plt.axvline(x = 0, color = 'black', linestyle='--') 

plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### La partie enti√®re inf√©rieure - plancher

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000) 
y = np.floor(x) 

plt.plot(x, y, label='y = |x|')  
plt.xlabel('x')  
plt.ylabel('|x|') 
plt.title('Graphique de la partie enti√®re inf√©rieure de x') 
plt.legend()  
plt.grid(True)  
plt.show()  
```

### La partie enti√®re sup√©rieure - plafond

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000) 
y = np.ceil(x) 

plt.plot(x, y, label='y = |x|')  
plt.xlabel('x')  
plt.ylabel('|x|') 
plt.title('Graphique de la partie enti√®re sup√©rieure de x') 
plt.legend()  
plt.grid(True)  
plt.show()  
```
