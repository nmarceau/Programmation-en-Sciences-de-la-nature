Le symbol ``%`` permet d'obtenir le reste d'une division entière. L'exercice consiste à écrire une fonction qui retourne la lettre suivante dans l'ordre alphabétique. La lettre qui suit ``z`` est ``a``.

```python
def lettre_suivante(lettre) :
    lettre = lettre.lower()
    return chr( ord("a") + (ord(lettre) - ord("a") + 1) % 26 )   

print(lettre_suivante("l"))
```

La fonction se termine après le premier ``return`` rencontré lors de l'exécution.

```python
def valeur_absolue(nombre):
    print("début")
    if nombre < 0: 
        y = -nombre
        return y
    else:
        y = nombre
        return y
    print("Jamais exécuté")
    
valeur_absolue(-150)
```

Une fonction peut être récursive : elle s'appelle elle-même. Mais il est important de savoir qu'il existe un cas dans lequel elle ne s'appelle pas pour arrêter la récursion.

```python
def recursive(nombre):
    if nombre / 2 < 1 :
        print(f"Fin nombre = {nombre}")
        return 1
    else:
        print(f"Appel de la fonction pour nombre = {nombre}")
        return recursive(nombre/2) + 1

recursive( 10 )
```

### Dictionnaires

#### clé - valeur

Une **liste** est un ensemble d'autres objets indexés par des entiers. Un **dictionnaire** est un ensemble d'autres objets indexés par presque n'importe quoi.

```python
dict = {} 
dict = {'couleur' : 'rouge', 'marque': 'toyota' } 

print(dict['couleur'])
print(len(dict))

marque = dict.get('marque')
print(marque)
```

On peut utiliser un dictionnaire pour compter les lettres d'un texte :

```python
texte = "exemple de texte"
dict = {}
for caractere in texte:
    dict[caractere] = dict.get(caractere,0) + 1

print(dict)
```
