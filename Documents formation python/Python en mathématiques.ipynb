## 1.2 Tracer un graphique



```python
import sympy as sym
from sympy.abc import x
from sympy.plotting import plot

y = 3 - x**2

plot(y)  # ou bien sym.plotting.plot(y)
```

Pour spécifier le domaine en $x$ :    

```python
plot(y, (x, -2, 2))
```

Pour aussi limiter y

```python
plot(y, (x, -2, 2), ylim=[-2,3])
```

Pour tracer plusieurs fonctions :    

```python
plot(y, y**2, (x, -3, 3))
```

On peut aussi mettre des domaines différents, ce qui permet de tracer des fonctions définies par morceaux :    

```python
plot((-x, (x, -2, 0)), (x, (x, 0, 2)))
```

Pour avoir la même couleur sur les deux morceaux, il faut séparer la commande plot en deux.

```python
p = plot(-x, (x, -2, 0), line_color="b", show=False)
p.extend(plot(x, (x, 0, 2), line_color="b", show=False))
p.show()
```

```python
plot(1/x, ylim=[-2,2])
```

# 3. Algèbre linéaire

Avec sympy

```python
import sympy as sym
from sympy.abc import x, y, z

eq1 = 2*x - 3*y + z - 3
eq2 = x - 3*z -7
eq3 = x - 2*y + 2*z - 4

sym.solve([eq1, eq2, eq3], x, y, z)
```
