# Les graphiques

## Exercice 1

Tracer le graphique de $f(x)=\cos(x)$, sur l'intervalle $[-2\pi, 2\pi]$.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 100)

plt.plot(x, np.cos(x))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

# Les courbes de tendance

## Exercice 1

En vous basant sur les données expérimentales ci-dessous et l'équation de la cinématique suivante :
$$h(t)=h_0+v_0 t-\frac{1}{2} gt^2$$
tracez une courbe de tendance quadratique pour déterminer $h_0$, $v_0$ et $g$.

```python
import numpy as np
import matplotlib.pyplot as plt

temps = np.array([0.0, 0.033, 0.067, 0.1, 0.133, 0.167, 0.2, 0.233, 0.267, 0.3, 0.333, 0.367, 0.4])
hauteur = np.array([1.03, 1.0, 0.97, 0.92, 0.86, 0.79, 0.71, 0.62, 0.52, 0.4, 0.28, 0.14, 0.0])

# Graphique des données expérimentales
plt.scatter(temps, hauteur, label="mesures")

# Courbe de tendance
z = np.polyfit(temps, hauteur, 2)
tendance = np.poly1d(z)

print(f"g = {-2*z[0]}")
print(f"v0 = {z[1]}")
print(f"h0 = {z[2]}")

plt.plot(temps, tendance(temps), label="tendance", color="red")

plt.xlabel("temps (s)")
plt.ylabel("hauteur (m)")
plt.legend()
plt.show()
```

## Exercice 2

Créer un ensemble de données expérimentales fictif en choisissant une fonction de votre choix, à laquelle vous ajoutez du bruit. Utiliser ensuite curve_fit() pour obtenir une courbe de tendance.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Données expérimentales fictives
n = 200
x_exp = np.linspace(-10, 10, n)
y_exp = 5.0*np.exp(-x_exp**2/2**2) + 0.1*(0.5 - np.random.random(n))

plt.scatter(x_exp, y_exp, label="expérimental")

# Courbe de tendance
def modele(x, A, x0):
    return A*np.exp(-x**2/x0**2)

z,cov = curve_fit(modele, x_exp, y_exp)
plt.plot(x_exp, modele(x_exp, *z), label="tendance", color = "r", linewidth=2)

incertitude = np.sqrt(np.diag(cov))

print(f"A = {z[0]:.3f} \u00B1 {incertitude[0]:.3}")
print(f"x0 = {z[1]:.3f} \u00B1 {incertitude[1]:.3}")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

# Méthodes numériques de base

## Exercice 1
Trouver les racines de l'équation $2x^2 + 2 - 1 = 0$ à l'aide de la méthode numérique de Newton.

***Il y a une erreur : utiliser plutôt l'équation $2x^2 + 2x - 1 = 0$***

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

# Nous voulons trouver les racines de la fonction suivante
def fonction(x):
  return 2*x**2 - 2*x - 1

# Exploration graphique pour estimer trouver les valeurs de départ
x = np.linspace(-1,2,100)
plt.plot(x, fonction(x))
plt.grid(True)

sol = newton(fonction, [0, 1])
print(sol)
```
