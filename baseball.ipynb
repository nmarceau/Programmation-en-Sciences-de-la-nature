```python
import math
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 25, 100)
h = -0.02*t**2 + 0.4*t

x1 = (-0.4 + math.sqrt(0.4**2 - 4*(-0.02)*0) ) / (2*(-0.02))
x2 = (-0.4 - math.sqrt(0.4**2 - 4*(-0.02)*0) ) / (2*(-0.02))

print(f"Les racines de l'équation quadratique sont {x1} et {x2}.")
print(f"La hauteur maximale est {max(h)} ou {-0.4**2/(4*-0.02)} mètres.") #sommet ( -b/2a, -(b^2 - 4ac)/4a )
print(f"La distance horizontale maximale est {abs(x2-x1)} mètres.")

plt.plot(t, h)
plt.axhline(y = 0, color = 'black', linestyle='--') 
plt.axvline(x = 0, color = 'black', linestyle='--') 
plt.ylim(0, 10);

plt.xlabel("distance horizontale (mètres)")
plt.ylabel("hauteur (mètres)")
plt.show()
```
