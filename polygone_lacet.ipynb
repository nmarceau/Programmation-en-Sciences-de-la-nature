# L'algorithme du lacet (Shoelace Algorithm)

L'algorithme du lacet est un algorithme mathématique permettant de déterminer l'aire d'un polygone dont les sommets sont décrits par leurs coordonnées cartésiennes dans le plan.

La méthode consiste à multiplier en croix les coordonnées correspondantes des différents sommets d'un polygone pour trouver son aire. Elle est appelée formule du lacet en raison de la multiplication croisée des coordonnées constituant le polygone, similaire au laçage des chaussures. (Voir le tableau ci-dessous).

https://accromath.uqam.ca/2024/02/la-formule-des-lacets-de-souliers/

Un cas d'utilisation de l'algorithme => https://adventofcode.com/2023/day/18

Lacet

$$A_{Polygone} = \frac{1}{2} \left\lvert  \sum_{i=1}^{n-1}x_iy_{i+1}+x_ny_1 - \sum_{i=1}^{n-1}{x_{i+1}y_i+x_1y_n}  \right\rvert $$

Pour appliquer l'algorithme du lacet, vous devez :

- Énumérer tous les sommets dans l'ordre anti-horaire (par exemple A, B, C, D, E) dans un tableau, et noter les coordonnées x et y dans deux colonnes distinctes du tableau.
- Calculer la somme en multipliant chaque coordonnée x par la coordonnée y dans la ligne suivante (en revenant à la première ligne lorsque vous atteignez la fin du tableau).
- Calculer la somme en multipliant chaque coordonnée y par la coordonnée x dans la ligne suivante (en revenant à la première ligne lorsque vous atteignez la fin du tableau).
- Soustraire la deuxième somme de la première et obtenir la valeur absolue de la différence |somme1 - somme2|.
- Diviser la valeur résultante par 2 pour obtenir l'aire du polygone.

```python
def calculer_aire_polygone(liste_sommets):
  
  nombre_sommets = len(liste_sommets)
  somme1 = somme2 = 0
  
  for i in range(0,nombre_sommets-1):
    somme1 = somme1 + liste_sommets[i][0] * liste_sommets[i+1][1]
    somme2 = somme2 + liste_sommets[i][1] * liste_sommets[i+1][0]
    
  somme1 = somme1 + liste_sommets[nombre_sommets-1][0] * liste_sommets[0][1]   
  somme2 = somme2 + liste_sommets[nombre_sommets-1][1] * liste_sommets[0][0]   
  
  return abs(somme1 - somme2) / 2
  
# Test 1; résultat attendu : 32 unités carrées  
A = [2,7]
B = [10,1]
C = [8,6]
D = [11,7]
E = [7,10]
polygone = [A,B,C,D,E]  
aire = calculer_aire_polygone(polygone)
print(f"Liste des sommets du polygone : {polygone}")
print(f"La surface du polygone est : {aire} unités carrées")
```
