On considère la fonction __f__ définie sur $\mathbb{R}$ par $\displaystyle f(x)=\frac{1}{4} x^3+x-3$.


Écrire une fonction nommée f qui :
* reçoit en argument une valeur $x$ ;
* renvoie son image par la fonction $f$.


```python
def f(x):
    return 1/4*x**3+x-3

print(f"f(0) = {f(0)}")
print(f"f(5) = {f(5)}")
```

Écrire une fonction nommée pente qui :
* reçoit en arguments les coordonnées de deux points $A(x_A;y_A )$ et $B(x_B;y_B )$ (avec $x_A≠x_B$) ;
* renvoie la pente de la droite $(AB)$.


```python
def pente(xA,yA,xB,yB):
    return (yB-yA)/(xB-xA)
pente(3,3,7,4)
```
