### Les fonctions trigonom√©triques

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 350, 800)

y1 = np.sin(np.radians(x))
plt.plot(x, y1, color='blue', label='sin(x)')

y2 = np.cos(np.radians(x))
plt.plot(x, y2, color='green', label='cos(x)')

y3 = np.tan(np.radians(x))
plt.plot(x, y3, color='red', label='tan(x)')

plt.xlabel('Abscisse x')
plt.ylabel('Ordonn√©e y')
plt.legend()

plt.ylim(-3,3)
plt.show()
```
