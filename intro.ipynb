## Introduction

Un langage de programmation permet de décrire avec précision des opérations très simples sur des données. Comme tout langage, il a une grammaire et des **mot-clés**. La complexité d'un programme vient de ce qu'il faut beaucoup d'opérations simples pour arriver à ses fins. Voyons cela quelques usages simples. Il vous suffit d'exécuter chaque petit extrait en appuyant sur le triangle pointant vers la droite ci-dessus. N'hésitez pas à modifier les extraits pour mieux comprendre ce que le programme fait.

**La calculatrice**

```python
a = 5
b = 10
c = a + b
print(c)    # affiche z
```

**L'addition, l'incrémentation**

```python
x = 3
y = x + 1
print(y)
x += 5
print(x)
```

**La répétition ou les boucles**

```python
somme = 0
for i in range (0, 10) :
    somme = somme + i             # répète dix fois cette ligne
print (somme)   
```

Le mot-clé ``print`` n'a pas d'incidence sur le programme. En revanche, il permet d'afficher l'état d'une variable au moment où on exécute l'instruction ``print``.

**Les conditions ou les tests**

```python
age = 18
if age >= 18:
    print("majeur")     # un seul des deux blocs est pris en considération
else:
    print("mineur")
```

**Les chaînes de caractères**

```python
nombre = 10
print(nombre)     # quelle est la différence
print("nombre")   # entre les deux lignes
chaine = "Sciences"
chaine += " informatique"
print(chaine)   
```

Toute valeur a un type et cela détermine les opérations qu'on peut faire dessus. ``2 + 2`` fait ``4`` pour tout le monde. ``2 + "2"`` fait quatre pour un humain, mais est incompréhensible pour l'ordinateur car on ajoute deux choses différentes (``nombre + texte``).

```python
print(2+2)
print("2" + "3")
```

**Une erreur de syntaxe**

```python
a = 0
for i in range (0, 10)   # il manque quelque chose sur cette ligne
    a = a + i
print(a)
```

**Une opération interdite**

```python
a = 0
s = "e"
print(a + s)    # impossible d'additionner un entier et une chaîne de caractères
```

+   Ecrire les instructions nécessaires pour calculer la somme des 10 premiers entiers au carré.

```python
somme = 0
for n in range(1, 11):
    somme += n**2
print(somme)  # affiche 385
```

+   Ecrire les instructions nécessaires pour calculer la somme des 5 premiers entiers impairs au carré.

```python
somme = 0
ctr = 1
for n in range(1, 6):
    somme += ctr
    ctr += 2
    
print(somme)  # affiche 25
```

+   Ecrire les instructions nécessaires pour calculer la somme des qui 10 premières factorielles : $\sum_{i=1}^{10} i!$.

```python
def factorielle(n):
    if n == 0:
        return 1
    else:
        return n * factorielle(n-1)

somme = 0
for i in range(1, 11):
    somme += factorielle(i)

print(f"La somme des factorielles de 1 à 10 est {somme}.")
```
