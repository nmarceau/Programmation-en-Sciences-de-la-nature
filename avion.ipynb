### Deux avions se faisant face se préparent à atterir sur la même piste. 

L'avion A est à une altitude de 3000 mètres. L'avion a déjà amorcé sa descente et elle se trouve à une distance du point de contact avec la piste de 9270 mètres. Sa vitesse est de 150 km/h.

L'avion B est également à une altitude de 3000 mètres. Elle amorcera sa descente dans 1000 mètres, puis sa descente jusqu'au point de contact avec la piste se trouvera alors à 6000 mètres au moment d'amorcer la descente. Sa vitesse est de 200 km/h.

- Quelle distance sépare les deux avions?
- Quelle est la mesure de l'angle de descente de chaque avion?
- Au moment où l'avion B touchera le sol, quelle sera l'altitude de l'avion A?

Avion

```python
import math
import matplotlib.pyplot as plt

angle_descente_A = math.asin(3000/9270)
print(f"L'angle de descente de l'avion A est {math.degrees(angle_descente_A):.2f}°")

angle_descente_B = math.asin(3000/6000)
print(f"L'angle de descente de l'avion B est {math.degrees(angle_descente_B):.2f}°")

#distance_A = 9270 * math.cos(angle_descente_A)
distance_A = math.sqrt(9270**2 - 3000**2)
print(f"La distance entre l'avion A et l'aéroport entre ligne droite est {distance_A:.2f} mmètres")

# distance_B = 6000 * math.cos(angle_descente_B)
distance_B = math.sqrt(6000**2 - 3000**2)
print(f"La distance entre l'avion B et l'aéroport entre ligne droite est {distance_B:.2f} mmètres")

distance_entre_avions = distance_A + distance_B + 1000
print(f"La distance entre les deux avions est {distance_entre_avions:.2f} mmètres")

altitude_A = math.sin(angle_descente_A) * (9270 - (7000*150/200))
print(f"L'altitude de l'avion A au moment où l'avion B touche le sol est {altitude_A:.2f} mètres")

xA = [0, -distance_A, -distance_A, 0]  
yA = [0, 0, 3000, 0]
plt.plot(xA, yA, label = "Trajectoire de l'avion A")


xB = [0, distance_B, distance_B, distance_B + 1000, distance_B, 0]
yB = [0, 0, 3000, 3000, 3000, 0]
plt.plot(xB, yB, label = "Trajectoire de l'avion B")

plt.axhline(y = 0, color = 'black', linestyle='--') 
plt.axvline(x = 0, color = 'black', linestyle='--') 
plt.ylim( 0, 10000)
plt.show()
```
