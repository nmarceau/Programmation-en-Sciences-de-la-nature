# Pandas et les dataframes

Pandas est une librairie faite pour le traitement de données avancé. Voici comment créer un dataframe simple

```python
import pandas as pd

df = pd.DataFrame({'Sexe' : ['F', 'F', 'M', 'M', 'M', 'F'], 'Moyenne':[83,80,75,64,85,80]})

print(df)
```

Pour lire un fichier csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")
```
