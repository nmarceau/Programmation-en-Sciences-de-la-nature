### Diagramme en bandes

```python
import matplotlib.pyplot as plt

parti = ['CAQ', 'PQ', 'LIB', 'PC', 'QS']
vote = [.35, .40, .10, .08, .07]

plt.title('Intentions de vote au Québec')
plt.bar(parti,vote, color=['deepskyblue','navy','red','blue','orange'], width=0.5)
plt.show()
```

### Diagramme en secteurs

```python
import matplotlib.pyplot as plt

parti = ['CAQ', 'PQ', 'LIB', 'PC', 'QS']
vote = [.35, .40, .10, .08, .07]

plt.pie(vote, labels = parti,      
           colors = ['deepskyblue','navy','red','blue','orange'],
           explode = [0, 0.2, 0, 0, 0],              
           autopct = lambda z: str(round(z, 2)) + '%', 
           pctdistance = 0.5,                    
           labeldistance = 1.1,                 
           normalize=False)                    

plt.title('Diagramme en secteurs')
plt.show()
```
