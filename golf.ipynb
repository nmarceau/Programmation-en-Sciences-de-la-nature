### La trajectoire d'une balle de golf

Au golf, l'analyse assistée par ordinateur permet d'étudier, entre autres, l'élan d'un golfeur ou d'une golfeuse, ou la trajectoire d'une balle d'après le type de bâton utilisé. Le schéma ci-dessous montre, pour trois types de fers, une partie de la trajectoire parabolique suivie par une balle de golf. Les mesures indiquées en rose, en vert et en orange correspondent aux hauteurs maximales atteintes par la balle.

Après avoir analysé chacune des trajectoires, recommandez le type de fer qui permettrait d'atteindre le vert en un seul coup.

golf

$$f(x) = a(x-h)^2+k$$
Les paramètres $h$ et $k$ représentent respectivement les coordonnées $x$ et $y$ du sommet.
Le paramètre $a$ permet de déterminer l'orientation (l'ouverture) de la parabole et la courbure de celle-ci.

```python
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(0, 200, 500)

# Fer 8
k = 47
h = 125 - 75
a = (22 - k) / h**2 # Utiliser le point (0, 22) pour trouver a
y_fer8 = a*(x-h)**2 + k
plt.plot(x, y_fer8, label = "Fer 8", color = "pink")

# Fer 7
k = 40
h = 125 - 65
a = (22 - k) / h**2
y_fer7 = a*(x-h)**2 + k
plt.plot(x, y_fer7, label = "Fer 7", color = "green")

# Forme générale de l'équation ; f(x) = -.005x^2 + 0.6x + 22; Pour trouver les points d'intersection, il faut résoudre l'équation f(x) = 5
# -.005x^2 + 0.6x + 22 = 5, # -.005x^2 + 0.6x + 17 = 0
# On résout par la formule quadratique; 
a = -.005
b = 0.6
c = 17
x1 = (-b + math.sqrt(b**2 - 4*a*c)) / (2*a)
x2 = (-b - math.sqrt(b**2 - 4*a*c)) / (2*a)
plt.plot(x2, 5, 'rx', label = "Point de contact avec le vert")

# Fer 6
k = 36.7
h = 125 - 55
a = (22 - k) / h**2
y_fer6 = a*(x-h)**2 + k
plt.plot(x, y_fer6, label = "Fer 6", color = "orange")

plt.axhline(y = 0, color = 'black', linestyle='--') 
plt.axvline(x = 0, color = 'black', linestyle='--') 
plt.plot([125,165], [5,5], color = 'green', linestyle = '--')
plt.legend()
plt.ylim(0, 50)
plt.xlim(0, 200)
plt.xlabel("x")
plt.ylabel("y")
plt.show()

print("Le graphique montre que le fer 7 est le meilleur choix pour atteindre le vert en un seul coup.")
```
