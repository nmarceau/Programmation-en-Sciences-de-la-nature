## La hauteur de trois pièces pyrotechniques en fonction du temps.

Chaque règle est de la forme canonique $$y = a(x - h)^2 + k$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 200)

h = 10
k = 250

a = -125/8
plt.plot(x, a*(x-h)**2 + k, label="a = -125/8", color="red")

a = -5/2
plt.plot(x, a*(x-h)**2 + k, label="a = -5/2", color="blue")

a = -125/2
plt.plot(x, a*(x-h)**2 + k, label="a = -125/2", color="green")

plt.ylim(0, 300);
plt.axhline(y = 0, color = 'black', linestyle='--') 
plt.axvline(x = 0, color = 'black', linestyle='--') 

plt.xlabel("temps écoulé depuis le lancement (s)")
plt.ylabel("hauteur (m)")
plt.legend()
plt.show()
```
