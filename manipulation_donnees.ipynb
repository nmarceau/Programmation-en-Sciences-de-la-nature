# Manipulation de données

## Numpy Arrays

```python
import numpy as np

tableau = np.random.random((3, 5))  # un tableau de 3x5 de valeurs aléatoires
print(tableau)
```

```python
print(tableau[0, 0]) # premier élément
print(tableau[1]) # deuxième ligne
print(tableau[:, 1]) # toutes les lignes, colonne 1
```

```python
print(tableau.T) # transposée
```

```python
y = np.linspace(0, 12, 5) # 5 valeurs entre 0 et 12
print(y)
```

```python
print(y[:, np.newaxis]) # transformation en vecteur colonne
```

## Scipy Sparse Matrices

```python
from scipy import sparse

tableau = np.random.random((10, 5)) # un tableau de 10x5 de valeurs aléatoires
print(tableau)
```

```python
tableau[tableau < 0.7] = 0 # on met à zéro les valeurs inférieures à 0.7
print(tableau)
```
